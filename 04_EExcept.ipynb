## MANEJO DE EXCEPCIÓN

```python
try:
    n1=int(input("Ingresa un numero"))
    n2=int(input("Ingresa un numero"))
    n3=n1/n2
    print(f"{n1}/{n2}={n3}")
except ValueError:
    print("No se permiten letras")
except ZeroDivisionError:
    print("Division entre 0")
except:
    print("Ocurrió una excepción")
else:
    print("No ocurrió ninguna excepción")
finally:
    print("ejecución terminada")
```
