```python
#crear una función que se llame hipotenusa que reciba cómo argumentos 2 catetos, y nos devuelva la hipotenusa
def hipotenusa(v1,v2):
    hipo = (v1**2+v2**2)**0.5
    return hipo
```

```python
hipotenusa(30,40)
```

```python
# crear una función que reciba n argumentos, y devuelva la suma de los mismos.
def sumaN(*x): #* - multiples argumentos
    sum=0
    for i in x:
        sum += i
    return sum
```

```python
sumaN(2,5)
```

```python
sumaN(1,7,8)
```

```python
# Implementar una función para calcular el valor de fibonacci
def fibo(n):
    if n==0 or n==1:
        return n
    else: 
        return fibo(n-1)+fibo(n-2)
```

```python
# sumar dos numeros diferentes de cero enteros, de manera recursiva permitiendose 
#solo sumar y restar unos
def suma(a,b):
    if b==0:
        return a
    else:
        return suma(a,b-1)+1
```

```python
suma(3,4)
```
