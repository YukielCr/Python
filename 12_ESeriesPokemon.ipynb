## Filtrado de una Serie

```python
materias=pd.Series({"matematicas":6,"economia":4.5,"Programasion":8.5})
```

```python
#Filtar aquellas materias cuya calificasion sea mayor a 5
print(materias[materias>5])
```

```python
# filtrar aquellaas materias donde sea igual a 6
print(materias[materias==6])
```

```python
print(materias[materias>=6].count())
```

```python
# Calcular la media de la calificasiones aprobatorias 
```

```python
print(materias[materias>=6].mean())
```

## Ordenar una serie

```python
materias
```

```python
## Ordenar las materias de acuerdo as sus valores acendentemente 
materias2=materias.sort_values()
materias2

#Por defecto dlo hace de manera acendente 
```

```python
# Ordenarlos de manera decendente 
materias3=materias.sort_values(ascending=False)
materias3
```
