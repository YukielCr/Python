## DICCIONARIOS

```python
# Un diccionario relaciona 2 valores: una llave y su valor. Es una colección de pares llave-valor (key-value)
# Las claves/llaves/key son inmutables.
# Los valores/value pueden cambiarse. 

d = {"nombre": "Pinocho", "edad":8, "pais":"Italia"}
d
```

```python
type(d)
```

```python
print(d["edad"])
print(d["pais"])
print(d["nombre"])
```

```python
# Renobrer
d["nombre"] = "mentiroso"
d
```

```python
# Manera 1 de obtener el nombre
d["nombre"]
```

```python
# Otra manera de obtener el name 
d.get("nombre")
```

```python
# eliminar un parametro de la lista 
del d["edad"]
```

```python
d
```

```python
# Checar si un campoesta en el diccionari
"nombre" in d
```

## Iterando Una lista

```python
equipos=["Kevin","Yukiel","Corona","Ramos"]
for name in equipos:
    print(name)
```

```python
nu=[10,20,30,40,50]
for n in nu:
    print(n)
```

```python
for k,v in enumerate(equipos,10):
    print(k," ",v)
```

```python
# Iterando una tupla
tupla=(10,"Kevin",10.5)
for t in tupla:
    print(type(t))
```

```python
#Iterando un diccionario 
D = {"nombre":"Gepeto","lugar":"Disneyland","profesion":"carpintero"}
for d in D:
    print(d)
```

```python
for d in D.values():
    print(d)
```

```python
for d in D:
    print(d," ",D[d])
```

```python
for d,v in D.items():
    print(d,v)
```

```python
#Iterando una cadena 
for men in "Hola mundo":
    print(men)
```

```python
#iterando una lista de valores con decimales y obteniendo una salida formateada
print("123456789")
valores = [1.5,3.45,89.678,9.36,2.3]
for i in valores:
    print(f"Valores: {i:2.2f}")
```
