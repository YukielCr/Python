## Inforasion traida desde una pagina Web

```python
import pandas as pd
import requests as rq
```

```python
respuesta=rq.get("https://huasteco.tiburcio.mx/marvel")
datos=respuesta.json()
print(datos)
```

```python
type(datos)
```

```python
datos[0]
```

```python
print(datos[0]["foto"])
```

```python
for i, dic in enumerate (datos):
    print(i,dic["nombreP"])
```

```python
# Escribe un programa en python 
# que muestre el nombre real, su descripsion y que que ponga una linea divisioria ------------ entre cada una

for i, dic in enumerate(datos):
    print(i,dic["nombreP"])
    print(" ",dic["nombreR"])
    print(" ",dic["descripcion"])
    print(" ",dic["foto"])
    print("------------------------------------------")
```

## Convertir una lista en un data Frame

```python
df=pd.DataFrame(datos)
df.head(3)
```

```python
# Definir nuestro propio index
df=df.set_index("id")
df
```

## Convertir un data Frame en un svs

```python
# Esto porque esta recibiendo los datos de la red
df.to_csv("csv/marvel2.csv")
```
