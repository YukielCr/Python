## Estructura condisional IF

```python
# Programa para saber si un alumno esta aprobado o reprobado

cal=int(input("Dame tu calificasion"))
if cal>=70:
    print("Aprobado")
else:
    print("Reprobado")
```

```python
#Tres posibles mensajes 
# temperatura menor a 0 "MUY FRIO"
# temperatura de los 0 a 15 "FRIO"
# temperatura de los 16 a 25 "Fresco
# temperatura mayor a 25 "Infierno"

t=int(input("Dime la temperatura"))

if t<0:
    print("Muy Frio")
elif t<=15:
    print("Frio")
elif t<=25:
    print("Fresco")
else:
    print("Infierno")
```

## CICLO WHILE

```python
# Imprimir numeros del 10 al 100
n=10
while n<=100:
    print(n)
    n+=10
```

```python
# Realiza la suma de los numeros capturados asta que se tecle el -1
suma=0
nu=int(input("Dame un numero"))
while nu != -1:
    if nu>=0:
        suma+=nu
    nu=int(input("Dame un numero"))
print(f"La suma es de {suma}")
```
