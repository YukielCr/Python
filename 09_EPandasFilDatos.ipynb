## Calcular nuevas columnas

```python
# Calcular un pago, que corresponda a las mensiones multiplicadas por 1.5 dolares
# Esto lo prondemos en una columna que se llama "PremioM"

def calcularPremio(m):
    return m*1.5

# Agregamos la columna premio y asignamos el valor 
df4limpio["PremioM"]=df4limpio["mentions"].apply(calcularPremio)
df4limpio.head(1)
```

```python
# Otra forma de Resolverlo 
df4limpio["PremioM"]=df4limpio["mentions"]*1.5
df4limpio.head(1)
```

```python
# Calcular la popularidad de cada fila en una columna
# Que se llame popularidad
# La popularidad se va a dividir los folloers entre followers

def popularida(fila):
    resultado=fila["followers"]/fila["followees"]
    return resultado

df4limpio["popularidad"]=df4limpio.apply(popularida,axis=1)
df4limpio.head(1)
```

```python
# Ejercicio: Calcular el promedio de favoritos por pilas 
df4_2=df4limpio[["favorites","country"]]
df4_2.groupby("country").mean
df4_2
```

```python
# Desviasion estandar
print(df4_2.groupby("country").mean().std)
# Medio
print(df4_2.groupby("country").mean().mean)
# maximo
print(df4_2.groupby("country").mean().max)
# minimo
print(df4_2.groupby("country").mean().min)
# variansa agrupada
print(df4_2.groupby("country").mean().var)
```
