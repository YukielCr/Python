```python
import pandas as pd
```

```python
df_A = pd.DataFrame([
    {"Nombre":"Mario","Perfil":"Electromecanico"},
    {"Nombre":"Sandra","Perfil":"Quimia"},
    {"Nombre":"Laura","Perfil":"Informatica"},
    {"Nombre":"Elsy","Perfil":"Quimica"}
])
df_A = df_A.set_index("Nombre")
df_A
```

```python
df_B = pd.DataFrame([
    {"Nombre":"Mario","Puesto":"Director"},
    {"Nombre":"Sandra","Puesto":"Supervisor"},
    {"Nombre":"Arturo","Puesto":"inspecto"},
    {"Nombre":"Laura","Puesto":"Director"}
])
df_B = df_B.set_index("Nombre")
df_B
```

```python
print(df_A)
print(df_B)
```

```python
## Marge Inner 
pd.merge(df_A,df_B,how="inner", 
         left_index=True,
         right_index=True)
```

## Merge outer

```python
## Marge Inner 
pd.merge(df_A,df_B,how="outer", 
         left_index=True,
         right_index=True)
```

## Marge right

```python
## Marge Inner 
pd.merge(df_A,df_B,how="right", 
         left_index=True,
         right_index=True)
```

```python
## Marge left
pd.merge(df_A,df_B,how="left", 
         left_index=True,
         right_index=True)
```
