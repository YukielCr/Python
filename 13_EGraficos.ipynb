# Graficos por pandas

```python
import pandas as pd
```

```python
df_population_raw=pd.read_csv("Python Tiburcio/population_total.csv")
```

```python
df_population_raw
```

## Hacer una tabla pivote

```python
# Bamos a reorganizar el dato de sep de poblasion total, de modo que obtengamos el nombre de los paises en las columnas y años y el indoce
# Esto no va a ayudar mas adelante cuando agamos la visualisasiones 

# Eliminar los valores nulos
df_population_raw.dropna(inplace=True)
```

```python
# Hacer una tabla pibote usando el metodo pivot(): paises en los nombres de las columnas y los años en el indice
df_pivot=df_population_raw.pivot(index="year",columns="country",values="population")
df_pivot
```

## Seleccionanos los paises Deseados

```python
df_pivot2=df_pivot[["United States","India","China","Indonesia","Brazil"]]
df_pivot2
```

# Graficos

## Graficos de lineas

```python
df_pivot2.plot(kind="line",
               xlabel="años",
               ylabel="poblasion",
               title="poblasion de 55 a 20",
               figsize=(6,3)
               )
```

## Grafico de Barras 

```python
# Seleccionanos solo el año 2020
df_pivot_2020=df_pivot2[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
# Trasformar el data frame(Cambiar filas por columnas y viseversa)
df_pivot_2020=df_pivot_2020.T
df_pivot_2020
```

```python
# Ahora si creamos la grafica 
df_pivot_2020.plot(kind="bar",
                   xlabel="paises",
                   ylabel="Poblasion",
                   title="Poblasion 2020",
                   color="blue"
                   )
```

## Grafico de pastel

```python
# Cambiar el nombre de de las columna 2020.0 por 2020
####### Duda
df_pivot_2020.rename(columns=({2020:"2020"}),inplace=True)
df_pivot_2020
```

```python
df_pivot_2020.plot(kind="pie",
                   y="2020",
                   title="Poblasion en 2020"
                   )
```
