## MANEJO DE ARCHIVOS DE TEXTO

```python
# Leectura de un archivo 
try:
    with open("mensaje.txt","r") as archivo:
        for m in archivo:
            print(m,end="")
except FileNotFoundError:
    print("el archivo no existe")
```

```python
# Escritura 
# Agregar una lista hasata el final 
try:
    with open("mensaje.txt","a") as archivo:
        archivo.write("Hola Munbdo \n")
except FileNotFoundError:
    print("el archivo no existe")
```

```python
# crear un archivo
# dado el siguiente diccionario, almacenarlo en un archivo de texto
calificaciones = {"Mario": 90,
                 "Elsie": 100,
                 "Leo" : 97,
                 "Arturo": 80,
                 "Laura": 95,
                 "Sandra": 100}
# al escribir en el archivo, hacerlo poniendo cada key-value en un reglón diferente, separados por comas
# tambien poner una cabecera en la primer linea del archivo: nombre, calificación.
with open ("calificaciones.txt","w") as archivo:
    archivo.write("nombre,calificacion\n")
    for nom, calif in calificaciones.items():
        archivo.write(f'{nom},{calif}\n')
    print("listo")
```
