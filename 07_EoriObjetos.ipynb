```python
import math as m
```

```python
class circulo:
    def __init__(self):
        self.radio=0

    def setRadio(self,valor):
        self.radio=valor

    def getRadio(self):
        return self.radio
    
    def area(self):
        return m.pi*m.pow(self.radio,2)
    
    def perimetro(self):
        return m.pi*self.radio*2
    
    def despliega(self):
        print(f'    Radio: {self.getRadio():8.4f}')
        print(f'Perimetro: {self.perimetro():8.4f}')
        print(f'     Area: {self.area():8.4f}')

     # metodo nuevo: areaSector
    def areaSector(self,ang):
        return ang*self.area()/360
```

```python
c1 = circulo()
c1.getRadio()
```

```python
c1.setRadio(10)
c1.getRadio()
```
